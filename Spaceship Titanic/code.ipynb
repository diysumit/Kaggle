Getting input features

```python
input_features = list(test_df.columns)
input_features.remove('PassengerId')
input_features.remove('Name')
input_features.remove('Cabin')
input_features
```

Binary values 1 and 0 shows wether a passenger was transported to next dimension or not               

```python
target_feature = train_df['Transported']
target_feature
```

## Random Forest Descision Tree Classifier

One hot encoding features

```python
X = pd.get_dummies(train_df[input_features])
X
```

```python
X_test = pd.get_dummies(test_df[input_features])
X_test
```

Handling null values in Datasets

```python
def check_null(df):
      print(df.isnull().any())
      if True in df.isnull().any():
            return True
      else:
            return False
```

```python
def impute(df: pd.DataFrame):
      columns = df.columns
      for column in columns:
            mean_value = df[column].mean()
            df[column].fillna(value=mean_value, inplace=True)
      print(df.isnull().sum())
      return df
```

```python
check_null(X)
check_null(X_test)
```

```python
impute(X)
impute(X_test)
```

## Creating Model

```python
model = RandomForestClassifier(n_estimators=100, max_depth=10, random_state=1)
model.fit(X, target_feature)
```

## Prediction

```python
predictions = model.predict(X_test)
predictions
```

## Saving output to CSV file

```python
output = pd.DataFrame({'PassengerId': test_df.PassengerId, 'Transported': predictions})
output
```

```python
output.to_csv('submission.csv', index=False)
```
